# Graph1 1: Distribution of Players by Country

```python

# Drop rows with missing values in 'flag' column
data_subset = data.dropna(subset=['flag'])

# Create a countplot for the distribution of players by country
plt.figure(figsize=(14, 6))
sns.countplot(x='flag', data=data_subset, palette='viridis')
plt.title('Distribution of Players by Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

# Analysis: Players are predemondily from the US in the Cardinals team in the last 5 years

# Graph 2:Experience Distribution of Players


```python
# Counts for each category in 'experience_range'
experience_range_counts = data['experience_range'].value_counts().reset_index()

#Column rename
experience_range_counts.columns = ['Experience Range', 'Count']

#Table
print("Count of Players in Each Experience Range:")
print(experience_range_counts)
```

```python
# Convert 'experience' to numeric, handling non-numeric values
data['experience'] = pd.to_numeric(data['experience'], errors='coerce')

# Define the bin edges and labels
bin_edges = [0, 5, 10, 15, 20, float('inf')]  # Adjusted ranges for a more even distribution
bin_labels = ['0-5', '6-10', '11-15', '16-20', '20+']

# Create a new column 'experience_range' with the assigned ranges
data['experience_range'] = pd.cut(data['experience'], bins=bin_edges, labels=bin_labels)

# Create the countplot for the experience ranges
sns.countplot(x='experience_range', data=data, palette='pastel')
plt.title('Count of Players in Each Experience Range')
plt.xlabel('Experience Range (Years)')
plt.ylabel('Count')
plt.show()
```

# Analysis: Most Cardinal Players in the last 5 years have experience less than 5 years in professional baseaball

# Graph 3:Scatterplot of Experience vs. Salary

```python

# Convert 'Salary' to numeric, handling non-numeric values
data['Salary'] = pd.to_numeric(data['Salary'].replace('[\$,]', '', regex=True), errors='coerce')

# Drop rows with missing values in 'Salary' or 'experience'
data = data.dropna(subset=['Salary', 'experience'])

# Create a scatter plot with color based on 'Salary'
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='experience', y='Salary', hue='Salary', data=data, palette='viridis', alpha=0.7)

# Lengenc for salary
scatter_plot.legend(title='Salary', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.title('Experience vs. Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')

# Format y-axis
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

# Analysis: Players with less than 5 years of experience recieve less salary compared to playes with 7.5 to 15 years

# Graph 4: Relationship of Games Played and Wins Above Replacement( Wins Player Added above the replaced Player) categorized based on thrwoing and batting

```python

# Drop rows with missing values in 'G_all', 'WAR', 'bats', 'throws' columns
data_subset = data.dropna(subset=['G_all', 'WAR', 'bats', 'throws'])

# Create a relplot
plt.figure(figsize=(14, 8))
sns.relplot(data=data_subset, x='G_all', y='WAR', col='throws', hue='bats', kind='scatter', palette='muted', height=6, aspect=0.8)
plt.suptitle('Relationship Between Games Played, WAR, and Batting/Throwing Styles', y=1.02)
plt.show()
```

# Analysis: Players with right arm throw who played more than 140 games and were Right side batter had a greate WAR. Players who are left arm thrower played lesser games and carried a less WAR. 

# Graph 5: Scatterplot of players games playes with bat and defennse and their weight


```python
# Drop rows with missing values in 'weight', 'G_batting', 'G_defense' columns
data_subset = data.dropna(subset=['weight', 'G_batting', 'G_defense'])

# Create a joint plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='G_batting', y='G_defense', hue='weight', data=data_subset, palette='viridis', alpha=0.7, size='weight')
plt.title('Relationship Between Weight and Games Batted vs. Games Played in Defense')
plt.xlabel('Games Batted')
plt.ylabel('Games Played in Defense')

# Add a legend
plt.legend(title='Weight', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
