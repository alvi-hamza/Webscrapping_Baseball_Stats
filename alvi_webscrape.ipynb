# Collect Player data from Multiple Pages (Final Data)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Function to get data from each year
def scrape_player_data_for_years(base_url, start_year, end_year):
    player_list = []

    # Loop through the range of years from start_year to end_year (inclusive)
    for year in range(start_year, end_year + 1):
        url = f"{base_url}{year}-roster.shtml#all_the40man"

        # Make a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract player data
            player_list.extend(extract_player_data_from_table(soup))
        else:
            print(f"Failed to retrieve data from {url}. Status code: {response.status_code}")

    return player_list

# Function to extract player data from the HTML table
def extract_player_data_from_table(soup):
    # Initialize an empty list to store player data dictionaries
    player_list = []

    # Find the table containing player statistics with the id 'div_appearances'
    stats_table = soup.find(id='div_appearances').tbody

    # Iterate through each row in the table
    for row in stats_table.find_all('tr'):
        # Extract individual player data fields from each row
        player = row.find(attrs={'data-stat': 'player', 'scope': 'row'}).text
        age = row.find(attrs={'data-stat': 'age'}).text
        flag = row.find(attrs={'data-stat': 'flag'}).text
        bats = row.find(attrs={'data-stat': 'bats'}).text
        throws = row.find(attrs={'data-stat': 'throws'}).text
        height = row.find(attrs={'data-stat': 'height'}).text
        weight = row.find(attrs={'data-stat': 'weight'}).text
        date_of_birth = row.find(attrs={'data-stat': 'date_of_birth'}).text
        experience = row.find(attrs={'data-stat': 'experience'}).text
        G_played = row.find(attrs={'data-stat': 'G_all'}).text
        GS = row.find(attrs={'data-stat': 'GS'}).text
        G_batting = row.find(attrs={'data-stat': 'G_batting'}).text
        G_defense = row.find(attrs={'data-stat': 'G_defense'}).text
        G_pitcher = row.find(attrs={'data-stat': 'G_p_app'}).text
        G_1b = row.find(attrs={'data-stat': 'G_1b'}).text
        G_2b = row.find(attrs={'data-stat': 'G_2b'}).text
        G_3b = row.find(attrs={'data-stat': 'G_3b'}).text
        G_ss = row.find(attrs={'data-stat': 'G_ss'}).text
        G_lf = row.find(attrs={'data-stat': 'G_lf_app'}).text
        G_cf = row.find(attrs={'data-stat': 'G_cf_app'}).text
        G_rf = row.find(attrs={'data-stat': 'G_rf_app'}).text
        G_outfield = row.find(attrs={'data-stat': 'G_of_app'}).text
        G_d_hitter = row.find(attrs={'data-stat': 'G_dh'}).text
        G_p_hitter = row.find(attrs={'data-stat': 'G_ph'}).text
        G_p_runner = row.find(attrs={'data-stat': 'G_pr'}).text
        WAR = row.find(attrs={'data-stat': 'WAR'}).text
        Salary = row.find(attrs={'data-stat': 'Salary'}).text
        Allstar_appearance = row.find(attrs={'data-stat': 'allstar_appearance'}).text

        # Append each player's data to the list as a dictionary
        player_list.append({
            'player': player, 'age': age, 'flag': flag,
            'bats': bats, 'throws': throws, 'height': height, 'weight': weight,
            'date_of_birth': date_of_birth, 'experience': experience,
            'G_all': G_played, 'GS': GS, 'G_batting': G_batting, 'G_defense': G_defense,
            'G_p_app': G_pitcher, 'G_1b': G_1b, 'G_2b': G_2b, 'G_3b': G_3b,
            'G_ss': G_ss, 'G_lf_app': G_lf, 'G_cf_app': G_cf, 'G_rf_app': G_rf,
            'G_of_app': G_outfield, 'G_dh': G_d_hitter, 'G_ph': G_p_hitter, 'G_pr': G_p_runner,
            'WAR': WAR, 'Salary': Salary, 'allstar_appearance': Allstar_appearance
        })

    # Return the list containing player data dictionaries
    return player_list

# Get data from each year
base_url = 'https://www.baseball-reference.com/teams/STL/'
start_year = 2019  # Adjust the starting year as needed
end_year = 2023    # Adjust the ending year as needed

# Send it to the function scrape_player_data_for_years
player_list_result = scrape_player_data_for_years(base_url, start_year, end_year)

# Print or process the scraped player data
print(player_list_result)

# Convert the list of player data to a DataFrame
player_data_final = pd.DataFrame(player_list_result)
```
